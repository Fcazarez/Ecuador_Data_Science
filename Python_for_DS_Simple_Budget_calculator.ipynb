Crear una calculadora de presupuesto personal que ayude a los usuarios a determinar los porcentajes de gastos para diferentes categorías, como salud, educación, alimentación, viajes y ahorro total.

Los estudiantes utilizarán conceptos básicos de entrada/salida, operaciones aritméticas y control de flujo en Python.



**Requisitos:**

El programa debe solicitar al usuario su sueldo mensual.
Calcular y mostrar los porcentajes de gastos para cada categoría.
Calcular y mostrar la cantidad de dinero destinada a cada categoría y al ahorro total.
Pasos Sugeridos:

Utilizar input() para obtener el sueldo mensual del usuario.
Determinar porcentajes razonables para cada categoría (salud, educación, alimentación, viajes y ahorro).
Calcular los montos correspondientes a cada categoría y al ahorro total.
Mostrar los resultados al usuario de manera organizada.

**Extras:**

Permitir al usuario ajustar los porcentajes de gastos para adaptarse a sus necesidades.
Mostrar los montos en una moneda específica y con formato de moneda.
Agregar la posibilidad de calcular los gastos anuales en lugar de mensuales.


```python
#IMport needed libraries, pandas to show results as dataframe, Locale to set up the mone amounts currency
import pandas as pd
```

```python
# Paso 1: Solicitar el sueldo mensual al usuario
while True:
    try:
        sueldo_mensual = float(input("Ingrese su sueldo mensual en su moneda local: "))
        break
    except ValueError:
        print("Por favor, ingrese un valor numérico válido.")
```

```python
#diccionario de rubros
rubros = {
    "Salud": 10,
    "Educación": 15,
    "Alimentación": 30,
    "Transporte": 10,
    "Entretenimiento": 5,
    "Vacaciones y Viajes": 10,
    "Ahorro Total": 20
}

# DataFrame a partir del diccionario
df = pd.DataFrame(list(rubros.items()), columns=["Rubro", "Porcentaje"])

# suma total de porcentajes
suma_total = df["Porcentaje"].sum()

# DataFrame para la fila "Total"
total_df = pd.DataFrame({"Rubro": ["Total"], "Porcentaje": [suma_total]})

# Concatenar el DataFrame de rubros y el DataFrame de "Total"
df = pd.concat([df, total_df], ignore_index=True)

# Mostrar el DataFrame
print("Rubros y porcentajes sugeridos:" )
df
```

```python
for rubro in rubros:
    while True:
        try:
            porcentaje = float(input(f"Ingrese el porcentaje para {rubro}: "))
            if 0 <= porcentaje <= 100:
                rubros[rubro] = porcentaje
                break
            else:
                print("El porcentaje debe estar entre 0 y 100.")
        except ValueError:
            print("Por favor, ingrese un valor numérico válido.")

# Verificar que la suma de porcentajes sea igual a 100
while sum(rubros.values()) != 100:
    print("La suma de los porcentajes no es igual a 100. Por favor, vuelva a ingresar los porcentajes.")
    for rubro in rubros:
        while True:
            try:
                porcentaje = float(input(f"Ingrese el porcentaje para {rubro}: "))
                if 0 <= porcentaje <= 100:
                    rubros[rubro] = porcentaje
                    break
                else:
                    print("El porcentaje debe estar entre 0 y 100.")
            except ValueError:
                print("Por favor, ingrese un valor numérico válido.")

```

```python
# Paso 3: Mostrar la tabla con rubros, porcentajes y cantidades mensuales
data = {"Rubro": rubros.keys(), "Porcentaje": rubros.values()}
print("Rubros y porcentajes actualizados:")
df = pd.DataFrame(data)
df
```

```python
# Calcular las cantidades mensuales
df["Monto Mensual ($)"] = (df["Porcentaje"] / 100) * sueldo_mensual
df
```

```python
# Paso 4: Solicitar al usuario si desea mostrar la información de forma anual
opcion = input("¿Desea mostrar la información de forma anual? (Sí/No): ").strip().lower()
if opcion == "si":
    # Calcular las cantidades anuales
    df["Monto Anual"] = df["Monto Mensual"] * 12
    print("\nPresupuesto Anual:")
    df1 =df[["Rubro", "Monto Anual"]].to_string(index=False, formatters={"Monto Anual": "${:,.2f}".format})
    df1
else:
    print("\nPresupuesto Mensual:")
    df1 = df[["Rubro", "Monto Mensual"]].to_string(index=False, formatters={"Monto Mensual": "${:,.2f}".format})
    df1

```
